```python
from itertools import chain
import json

from sklearn import preprocessing
import numpy as np

import pandas as pd

new_brain_data_array = []

# LOAD JSON FILE
braindata = pd.read_json('DoubleNewData.json')

# GRAB ARRAY OF TIMESTAMPS
brain_data_array = braindata['AllData']

timestamp_iterator = 0

# NORMALIZE Z VALUE FOR EACH TIMESTAMP
for timestamp in brain_data_array:
    timestamp_data = timestamp['Data']
    print(timestamp_data)
    x = timestamp_data[::3]
    # print(x)
    y = timestamp_data[1::3]
    # print(y)
    z = timestamp_data[2::3]
    
    #look at map function? 
    
    # NORMALIZE Z
    # 1) turn into np array type
    np_array = [np.array(z)]
    # 2) normalize using sklearn
    normalized_np_array = preprocessing.normalize(np_array)
    z = normalized_np_array.tolist()[0]
    
    #scale the Z data
    #Z_scale = np.array([[ 0.005, 0.05,  0.01]])
    #z = preprocessing.scale(Z_scale)

    # recombine data
    # final_list = zip(x, y, z)
    # final_list = [list(a) for a in zip(x, y, z)]
    final_modified_timestamp = [i for j in zip(x,y,z) for i in j]
    print('final_list')
    
    new_brain_data_object = {'Timestamp': timestamp_iterator, 'Data': final_modified_timestamp}
    new_brain_data_array.append(new_brain_data_object)
    
    timestamp_iterator = timestamp_iterator + 1
    
print(new_brain_data_array)

# CONVERT BACK INTO JSON

with open('new_braincoordinates.json', 'w') as json_file:
  json.dump(new_brain_data_array, json_file)
```
