```python
from itertools import chain
import json

with open('C:/Users/natas/Downloads/Json Brain Randomizer/MNI_Coordinate_Match.json', 'r') as f:
    braindata = json.load(f)

   
braindict = braindata['AllData'][0]

newbraindict = {k: v for k, v in braindict.items() if 'Data' in k} 

timedict = {k: v for k, v in braindict.items() if 'TimeStamp' in k} 

newbrainlist = [n for n in newbraindict.values()][0]

coordinatesx = [x for x in newbrainlist[::3]]
coordinatesy = [y for y in newbrainlist[1::3]]
coordinatesz = [z for z in newbrainlist[2::3]]

x = list(map(lambda x1, x2: x1 + x2, coordinatesx[::2], coordinatesx[1::2]))
y = list(map(lambda y1, y2: y1 + y2, coordinatesy[::2], coordinatesy[1::2]))
z = list(map(lambda z1, z2: z1 + z2, coordinatesz[::2], coordinatesz[1::2]))

listtupledata = list(zip(x, y, z))

# listlistdata = [list(i) for i in listtupledata]

newdata = list(chain.from_iterable(listtupledata))

# datadict = dict(zip([k for k in newbraindict.keys()], newdata))

braindatadict = dict(zip([k for k in newbraindict.keys()], [newdata]))  # newdata must be in [] to be read as a sequence

timedict.update(braindatadict)

valueslist = [timedict]

 neobraindata = dict(zip(braindata.keys(), [valueslist]))  # new coordinates ready to import as json
```

```python
with open('Neobraincoordinates.json', 'w') as json_file:
  json.dump(neobraindata, json_file)
```
